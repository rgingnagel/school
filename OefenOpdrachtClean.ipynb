```python
from IPython import display
import os
def show_app(app, port = 9999, 
             width = 700, 
             height = 350, 
             offline = False,
            in_binder = None):
    in_binder ='JUPYTERHUB_SERVICE_PREFIX' in os.environ if in_binder is None else in_binder
    if in_binder:
        base_prefix = '{}proxy/{}/'.format(os.environ['JUPYTERHUB_SERVICE_PREFIX'], port)
        url = 'https://hub.mybinder.org{}'.format(base_prefix)
        app.config.requests_pathname_prefix = base_prefix
    else:
        url = 'http://localhost:%d' % port
        
    iframe = '<a href="{url}" target="_new">Open in new window</a><hr><iframe src="{url}" width={width} height={height}></iframe>'.format(url = url, 
                                                                                  width = width, 
                                                                                  height = height)
    
    display.display_html(iframe, raw = True)
    if offline:
        app.css.config.serve_locally = True
        app.scripts.config.serve_locally = True
    return app.run_server(debug=False, # needs to be false in Jupyter
                          host = '0.0.0.0',
                          port=port)
```

```python
import plotly.figure_factory as ff

app = dash.Dash()
app.layout = html.Div([
    dcc.Dropdown(
        id='my-dropdown',
            options=[{'label':name, 'value':name} for name in lt[2].unique()],
        value='amsterdam'
    ),
    html.Div(id='my-graph'),
    html.Div(id='my-graph2'),
    dcc.Graph(id='my-table'),
])


@app.callback(
    dash.dependencies.Output('my-table', 'figure'),
    [dash.dependencies.Input('my-dropdown', 'value')])
def update_table(value):
    lt_value=lt[lt[2].str.contains(value)]
    lt_not_value=lt[~lt[2].str.contains(value)]
    value_vs_rest=pd.DataFrame([lt_not_value.groupby(5)[4].mean(),lt_value.groupby(5)[4].mean()] ).T
    value_vs_rest.columns=['rest', value]
    value_vs_rest['Oplossing'] = value_vs_rest.index
    value_vs_rest.sort_values(value, ascending=False)
    new_table_figure = ff.create_table(value_vs_rest)
    return new_table_figure

@app.callback(
    dash.dependencies.Output('my-graph', 'children'),
    [dash.dependencies.Input('my-dropdown', 'value')])
def update_graph(value):
    lt_value=lt[lt[2].str.contains(value)]
    lt_not_value=lt[~lt[2].str.contains(value)]
    value_vs_rest=pd.DataFrame([lt_not_value.groupby(5)[4].sum(),lt_value.groupby(5)[4].sum()] ).T
    value_vs_rest.columns=['rest', value]
    value_vs_rest.sort_values(value, ascending=False)

    return dcc.Graph(
    id='myGraph',
    figure={
        'data':[
            {'y': value_vs_rest.index, 'x': (value_vs_rest/value_vs_rest.sum()).sort_values(value)[value], 'type': 'bar', 'name': value, 'orientation': 'h'},
            {'y': value_vs_rest.index, 'x': (value_vs_rest/value_vs_rest.sum()).sort_values(value)["rest"], 'type': 'bar', 'name': "rest", 'orientation': 'h'},
        ],
        'layout': { 
            'title': "Percentage oplossingen per kind in een uitgevallen klas in " + value + " en in de rest van Nederland",
            'margin' : {'l': 230, 'b': 40, 't': 50, 'r': 5}
        }
    }
    )

@app.callback(
    dash.dependencies.Output('my-graph2', 'children'),
    [dash.dependencies.Input('my-dropdown', 'value')])
def update_graph(value):
    lt_ams=lt[lt[2].str.contains(value)]
    lt_not_ams=lt[~lt[2].str.contains(value)]
    ams_vs_rest=pd.DataFrame([lt_not_ams[5].value_counts(),lt_ams[5].value_counts()] ).T
    ams_vs_rest.columns=['rest', value]
    ams_vs_rest.index.name='Oplossing'
    ams_vs_rest.sort_values(value, ascending=False)

    return dcc.Graph(
    id='myGraph2',
    figure={
        'data':[
            {'y': ams_vs_rest.index, 'x': (ams_vs_rest/ams_vs_rest.sum()).sort_values(value)[value], 'type': 'bar', 'name': value, 'orientation': 'h'},
            {'y': ams_vs_rest.index, 'x': (ams_vs_rest/ams_vs_rest.sum()).sort_values(value)["rest"], 'type': 'bar', 'name': "rest", 'orientation': 'h'},
        ],
        'layout': {
            'title': "Percentage oplossingen per uitgevallen klas in " + value + " en in de rest van Nederland" ,
            'margin' : {'l': 230, 'b': 40, 't': 50, 'r': 5}
        }
    }
    )
```

```python
from geopy.exc import GeocoderTimedOut
geolocator = Nominatim()

def do_geocode(geoString):
    try:
        return geolocator.geocode(geoString)
    except GeocoderTimedOut:
        return do_geocode(geoString)

for city in cities:
    geoString = city + ", Nederland"
    location = do_geocode(geoString)
    if location:
        cityLatLong.loc[city].lat = location.latitude
        cityLatLong.loc[city].long = location.longitude 
    else:
        print(city)
    
    
```
